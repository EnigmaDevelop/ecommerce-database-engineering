```python
# Re-install package within clean session if needed
!pip install kafka-python-ng

import time
import json
import random
from datetime import datetime
from kafka import KafkaProducer

# CRITICAL FIX: Changing connection string from localhost to the container network alias 'kafka'
# and using the mapped intra-network port 29092
KAFKA_BROKER = "kafka:29092"
TOPIC_NAME = "clickstream_events"

# Configuration vectors for dynamic e-commerce telemetry simulation
URL_POOL = [
    "https://ecommerce.com",
    "https://ecommerce.com",
    "https://ecommerce.com",
    "https://ecommerce.com",
    "https://ecommerce.com"
]
DEVICE_POOL = ["desktop", "mobile_ios", "mobile_android", "tablet"]

print("🔌 Establishing a continuous link with the local Apache Kafka Message Broker...")

# Hardening connection setup by setting api_version to bypass handshake delays inside Docker network
producer = KafkaProducer(
    bootstrap_servers=[KAFKA_BROKER],
    api_version=(7, 5, 0),
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)
print("✅ Link secure! Streaming event loop initialized.\n")

# Start an infinite loop to simulate production traffic velocity
try:
    message_counter = 1
    while True:
        # Programmatically synthesize realistic multi-tenant transaction metrics
        event_payload = {
            "user_id": random.randint(1000, 50000),
            "page_url": random.choice(URL_POOL),
            "ip_address": f"192.168.1.{random.randint(2, 254)}",
            "device_type": random.choice(DEVICE_POOL),
            "click_timestamp": datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        }
        
        # Ship payload directly over the network partition array
        producer.send(TOPIC_NAME, value=event_payload)
        
        print(f"[{message_counter}] Event pushed onto broker -> User {event_payload['user_id']} visited {event_payload['page_url'].split('/')[-1]}")
        
        message_counter += 1
        time.sleep(0.5) # Throttled frequency (2 messages per second)
        
except KeyboardInterrupt:
    print("\n🛑 Event injection safely interrupted by engineer.")
finally:
    producer.flush()
    producer.close()
```
